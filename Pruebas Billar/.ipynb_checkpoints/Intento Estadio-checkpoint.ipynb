```julia
using PyPlot
```

```julia
posBarrx1=5.0 #delimitamos al rectangulo
posBarrx2=-5.0 
posBarry1=4.0
posBarry2=-4.0

err=0.25
it=0  #iteracion

posx=zeros(1000); #posicion en x
posy=zeros(1000);#posicion en y
x0=0.;#posicion en x         ya se definio
y0=0.;#posicion en y         Ya se definieron las dos
vy=1
vx=1
t=linspace(0,10,1000);



for j in 1:3
    
   for i in 1:1000
    x=x0+vx*t[i]
    y=y0+vy*t[i]
    posx[i]=x;
    posy[i]=y; 
    end

        for i in 1:1000
            
            ## Choque pared costados
            if abs(abs(posx[i])-posBarrx1)<err && abs(posy[i])< posBarry1 #abs(abs(posx[i])-posBarrx1) < abs(abs(posy[i])-posBarry1)
            #println(j,"        ",posx[i],"             ",posy[i],"             ", t[i],"      a     ", abs(posx[i]-posBarrx1)) 
            it=i #iteracion en la que chocan
            vx=-vx
            
            ## Choque pared arriba
            elseif abs(abs(posy[i])-posBarry1)<err && abs(posx[i])<posBarrx1 #abs(abs(posy[i])-posBarry1) < abs(abs(posx[i])-posBarrx1)<err
            #println(j,"        ",posx[i],"             ",posy[i],"             ", t[i],"       b     ", abs(posy[i]-posBarry1)) 
            it=i #iteracion en la que chocan
            vy=-1*vy            
        end
    end
   
x0=posx[it]
y0=posy[it]

plot(posx, posy)
grid("on")
plot([-5,5],[-4,-4],linestyle="-",linewidth=.8)
plot([-5,5],[4,4],linestyle="-",linewidth=.8)
plot([-5,-5],[-4,4],linestyle="-",linewidth=.8)
plot([5,5],[-4,4],linestyle="-",linewidth=.8)
xlim(-10, 10)
ylim(-10, 10)
end
```
