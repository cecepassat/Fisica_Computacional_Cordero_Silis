```julia
## Definimos la función que representa la derivada
H(x,t)= -2.5x;
## También necesitamos su derivada respecto a x
dxH(x,t)=-2.5;
```

```julia
## Para una condición inicial x(0)=0

## Para h=1
listt1=0:1:8
## Para h=0.5
listt2=0:0.5:8
```

```julia
##### Euler Implícito

SOL1=herramientas.Imp_Euler(H,dxH,listt1,1)
SOL2=herramientas.Imp_Euler(H,dxH,listt2,0.5)
```

```julia
using PyPlot
```

```julia
plot(listt1,SOL1,label="Solución numérica h=1")
plot(listt2,SOL2,label="Solución numérica h=0.5")
plot(linspace(0,8,100),e.^(-2.5(linspace(0,8,100))),label="Solución analítica");

xlabel("t");
ylabel("x(t)");
legend();
show();
```

Para _Punto Medio_ y _Runge-Kutta Orden 4_ resolvamos la siguiente ecuación:

$$\frac{dx}{dt}=\frac{2x-18 t}{1+t}.$$

Cuya solución analítica, sabemos que es $x(t)=4+8t-5t^2$, cuando $x(0)=4$.
Además escogemos $t\in(0,12)$

```julia
## Definimos la función que nos da la derivada
w(x,t)=(2x-18t)/(1+t)
```

```julia
## Definimos los arreglos de t's

## Para h=0.5
timelist1=0:0.5:12
## Para h=0.2
timelist2=0:0.2:12
```

```julia
##### Punto Medio

SOLPM1=herramientas.Punto_M(w,timelist1,4.0);
SOLPM2=herramientas.Punto_M(w,timelist2,4.0);
```

```julia
## Graficamos las soluciones para las distintas h's

plot(timelist1,SOLPM1,label="h=0.5");
plot(timelist2,SOLPM2,label="h=0.2");

list=linspace(0,12,100)
plot(list,4+8*list-5*list.^(2),label="Solución analítica")     ## Solución analítica
xlabel("t");
ylabel("x(t)");


legend();
show();
```

```julia
##### Runge-Kutta orden 4

## Para h=0.5
SOLRK1=herramientas.RK_O4(w,timelist1,4.0);
## Para h=0.2
SOLRK2=herramientas.RK_O4(w,timelist2,4.0);
```

```julia
## Graficamos las soluciones para las distintas h's

plot(timelist1,SOLRK1,label="h=0.5");
plot(timelist2,SOLRK2,label="h=0.2");


## Graficamos la solución analítica
list=linspace(0,12,100)
plot(list,4+8*list-5*list.^(2),label="Solución analítica");
xlabel("t");
ylabel("x(t)");


legend();
show();
```
