```julia
posBarrx1=5 #deilimitamos al rectangulo
posBarrx2=-5 
posBarry1=4
posBarry2=-4

err=0.05
it=0  #iteracion

posx=zeros(1000); #posicion en x
posy=zeros(1000);#posicion en y
x0=0.;#posicion en x         ya se definio
y0=0.;#posicion en y         Ya se definieron las dos
vx=4
vy=-2
t=linspace(0,10,1000);

for j in 1:3
    
   for i in 1:1000
    x=x0+vx*t[i]
    y=y0+vy*t[i] 
    posx[i]=x;
    posy[i]=y; 
    end 
    
        for i in 1:1000
            
        ## Choque en pared derecha-arriba
            if abs(posx[i]-posBarrx1)<err && abs(posy[i]) < abs(posx[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vx=-vx
            print("a")
        
        ## Choque en pared arriba-derecha
            elseif abs(posy[i]-posBarry1)<err && abs(posx[i]) < abs(posy[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vy=-vy
            print("b")
            
        ## Choque en pared arriba-izquierda
            elseif abs(posy[i]-posBarry1)<err && abs(posx[i]) < abs(posy[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vy=-vy  
            print("c")
    
        ## Choque en pared izquierda-arriba
            elseif abs(posx[i]-posBarrx2)<err && abs(posy[i]) < abs(posx[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vx=-vx
            print("d")
            
         ## Choque en pared abajo-izquierda
            elseif abs(posy[i]-posBarry2)<err && abs(posx[i]) < abs(posy[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vy=-vy  
            print("e")
            
        ## Choque en pared derecha-abajo
            elseif abs(posx[i]-posBarrx1)<err && abs(posy[i]) < abs(posx[i])
            println(posx[i],"             ",posy[i],"             ", t[i]) 
            it=i #iteracion en la que chocan
            vx=-vx
            print("f")
        
        #elseif println("no")
            end
       
        
        
        end
    
    
    
    
x0=posx[it]
y0=posy[it]

plot(posx, posy)
grid("on")
    plot([-5,5],[-4,-4],linestyle="-",linewidth=2.0)
plot([-5,5],[4,4],linestyle="-",linewidth=2.0)
plot([-5,-5],[-4,4],linestyle="-",linewidth=2.0)
plot([5,5],[-4,4],linestyle="-",linewidth=2.0)
end
```
